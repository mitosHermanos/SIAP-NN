```python
from nltk.corpus import stopwords 
from nltk.stem.wordnet import WordNetLemmatizer
import string
# nltk.download('wordnet
stop = stopwords.words('english')
# print(stop)
# import spacy
# nlp = spacy.load('en_core_web_lg')
stop.extend(['nothing','wasnt','could', 'would','didnt', 'didn\'t', 'wasn\'t', 'everything', 'room', 'one', 'hotel', 'u'])
# print(stop)
exclude = set(string.punctuation) 
lemma = WordNetLemmatizer()
def clean(doc):
#     [(x, y) for x in [1,2,3] for y in [3,1,4] if x != y]
#     print(doc.lower().split())
#     for ch in doc.lower().split():
#         print(ch)
# [' '.join(c for c in s.lower().split() if c not in stop_words) for s in q2_p ]
    doc = doc.replace('’', '\'')
    doc = doc.replace('/', ' ')
    doc = doc.replace('-', ' ')
    doc = doc.replace(',', ' ')
    doc = doc.replace('.', ' ')

    punc_free = [''.join(i for i in ch if not i in exclude) for ch in doc.lower().split()]
    stop_free = " ".join([i for i in punc_free if i.lower() not in stop])

    normalized = " ".join(lemma.lemmatize(word) for word in stop_free.split())
#     print('d',normalized)

    normalized = " ".join([i for i in normalized.split() if i.lower() not in stop])

    return normalized
```

```python
from nltk.corpus import stopwords 
from nltk.stem.wordnet import WordNetLemmatizer
import string
# nltk.download('wordnet
stop = stopwords.words('english')
# print(stop)
# import spacy
# nlp = spacy.load('en_core_web_lg')
stop.extend(['nothing','wasnt','could', 'would','didnt', 'didn\'t', 'wasn\'t', 'everything', 'room', 'one', 'hotel', 'u'])
# print(stop)
exclude = set(string.punctuation) 
lemma = WordNetLemmatizer()
def clean(doc):
#     [(x, y) for x in [1,2,3] for y in [3,1,4] if x != y]
#     print(doc.lower().split())
#     for ch in doc.lower().split():
#         print(ch)
# [' '.join(c for c in s.lower().split() if c not in stop_words) for s in q2_p ]
    punc_free = [''.join(i for i in ch if not i in exclude) for ch in doc.lower().split()]
    stop_free = " ".join([i for i in punc_free if i.lower() not in stop])

    normalized = " ".join(lemma.lemmatize(word) for word in stop_free.split())
#     print('d',normalized)

    normalized = " ".join([i for i in normalized.split() if i.lower() not in stop])

    return normalized

doc_clean = [clean(doc).split() for doc in sentences]  
# print(doc_clean)

# Creating the term dictionary of our courpus, where every unique term is assigned an index. 
dictionary = corpora.Dictionary(doc_clean)
dictionary.filter_extremes(no_below=15, no_above=0.8)
# print(dictionary)
# Converting list of documents (corpus) into Document Term Matrix using dictionary prepared above.
doc_term_matrix = [dictionary.doc2bow(doc) for doc in doc_clean]
# print(doc_term_matrix)

# Creating the object for LDA model using gensim library
Lda = gensim.models.ldamodel.LdaModel

# Running and Trainign LDA model on the document term matrix.
ldamodel = Lda(doc_term_matrix, num_topics=100, alpha=0.001, id2word = dictionary, passes=20)
# print(ldamodel.print_topics(num_topics=100))
```

```python
# corpus = []
# for i in np.array(list(embeddings.values())):
#     corpus.append(i)
# #     print(i)
# b_tree = ball_tree(np.array(corpus))

def get_result(str_name):
    df = pd.read_csv('hotels-parsed/' + str_name)
    df['Nights_stayed'] = df['Nights_stayed'].str.replace('nights', '')
    df['Nights_stayed'] = df['Nights_stayed'].str.replace('night', '')
    df['Negative'] = 0
    df['Positive'] = 0
    del df['Unnamed: 0']

    new_df = pd.DataFrame(columns=df.columns)

    new_df = new_df.drop(columns=['Negative_reviews'])
    new_df = new_df.rename(columns={"Positive_reviews": "Review"})
    df_neg = df.drop(columns=['Positive_reviews'])

    for index, row in df_neg.iterrows():
        if not pd.isna(row['Negative_reviews']):
            df_neg.loc[index,"Negative"]=1

    df_neg = df_neg.rename(columns={"Negative_reviews": "Review"})

    df_neg = df_neg.dropna()
    df_neg = df_neg.drop(index=df_neg[df_neg['Review'].str.lower().str.strip() == ('n/a' or 'na')].index)
    # df_neg = df_neg[~(df_neg['Review'].str.lower().replace('\xa0', '').replace(' ', '') == 'n/a')]
    print("NEG: ", len(df_neg.index))
    df_pos = df.drop(columns=['Negative_reviews'])

    for index, row in df_pos.iterrows():
        if not pd.isna(row['Positive_reviews']):
            df_pos.loc[index,"Positive"]=1

    df_pos = df_pos.rename(columns={"Positive_reviews": "Review"})

    df_pos = df_pos.drop(index=df_pos[df_pos['Review'].str.lower().str.strip() == ('n/a' or 'na')].index)
    

    df_pos = df_pos.dropna()
    print("POS: ", len(df_pos.index))
    
    
    final_df = df_pos.append(df_neg, ignore_index=True)
#     final_df.Review
    final_df['Facilities'] = ''
    final_df = final_df.drop_duplicates(keep='first', ignore_index=True)
    
    final_df.head()
    reviews = []
    for r in final_df.Review:
        reviews.append(clean(r))
    print("FINAL: ", len(reviews))
    return reviews, final_df
```
